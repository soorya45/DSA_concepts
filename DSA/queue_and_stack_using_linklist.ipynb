```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
class l:
    def __init__(self):
        self.head=None

    def push(self,data):
        n=Node(data,self.head)
        self.head=n
    def pop(self):
        if self.head is None:
            print("empty")
        else:
            t=self.head
            self.head=t.next
            print("deleted item:",t.data)
            t=None
    def display(self):
        if self.head is None:
            print("empty")
        else:
            monkey=self.head
            while monkey:
                print(monkey.data)
                monkey=monkey.next
    def peek(self):
        if self.head is None:
            print("empty")
        else:
            print(self.head.data)

o=l()
while True:
    a=int(input("push\npop\ndisplay\npeek\nenter ur choice"))
    if a==1:
        o.push(int(input("data")))
    elif a==2:
        o.pop()
    elif a==3:
        o.display()
    elif a==4:
        o.peek()

    else:
        break
```

```python
class Queue():

    def __init__(self, limit = 5):
        self.queue = []
        self.max = limit
        self.front = 0
        self.rear = -1



    def enqueue(self, item):
        if self.rear == self.max - 1:
          print("Queue is Full")

        else:
          if self.front == -1:
            self.front +=1
          self.rear = (self.rear + 1)
          self.queue.append(item)


    def dequeue(self):
          if self.rear == -1:
            print("Queue is empty")
          else:
            print("Deleted item:",self.queue[self.front])
            self.queue[self.front]=None
            self.front = self.front +1

    def status(self):
        if self.rear == -1:
          print("Queue is Empty")

        else:
          i = self.front
          while i <= self.rear:
            print(self.queue[i])
            i += 1

    def Rear(self):
        if self.rear == -1:
          print("Queue is Empty")

        else:
          print(self.queue[self.rear])

q = Queue()
while True:
        a=int(input('\n1. Enqueue \n2. Dequeue \n4. Rear \n5. Status \n6. Exit\nEnter your option:'))

        if a == 1:
            q.enqueue(input('Enter item: '))
        elif a == 2:
            print(q.dequeue())
        elif a == 4:
           print(q.Rear())
        elif a == 5:
            q.status()
        else:
            break;
```
