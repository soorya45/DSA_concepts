```python
class Queue():

    def __init__(self, limit = 5):
        self.queue = []
        self.max = limit
        self.front = 0
        self.rear = -1



    def enqueue(self, item):
        if self.rear == self.max - 1:
          print("Queue is Full")

        else:
          if self.front == -1:
            self.front +=1
          self.rear = (self.rear + 1)
          self.queue.append(item)


    def dequeue(self):
          if self.rear == -1:
            print("Queue is empty")
          else:
            print("Deleted item:",self.queue[self.front])
            self.queue[self.front]=None
            self.front = self.front +1

    def status(self):
        if self.rear == -1:
          print("Queue is Empty")

        else:
          i = self.front
          while i <= self.rear:
            print(self.queue[i])
            i += 1

    def Rear(self):
        if self.rear == -1:
          print("Queue is Empty")

        else:
          print(self.queue[self.rear])

q = Queue()
while True:
        a=int(input('\n1. Enqueue \n2. Dequeue \n4. Rear \n5. Status \n6. Exit\nEnter your option:'))

        if a == 1:
            q.enqueue(input('Enter item: '))
        elif a == 2:
            print(q.dequeue())
        elif a == 4:
           print(q.Rear())
        elif a == 5:
            q.status()
        else:
            break;

```
